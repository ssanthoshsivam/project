```python
#Join query
cursors.execute("""select a.id,b.neo_reference_id from asteriods a
join close_approach b on a.id=b.neo_reference_id""")
cursors.fetchall()
```

```python
#1. count how many times each asteriod has approched earth
cursors.execute("select neo_reference_id,count(*) from close_approach group by neo_reference_id")
cursors.fetchall()
```

```python
#2. average velocity of each asteriod over multiple approches
cursors.execute("select neo_reference_id,avg(relative_velocity_kmph) from close_approach group by neo_reference_id")
cursors.fetchall()
```

```python
#3. list top 10 fastest asteriods

cursors.execute("select a.id,a.name,b.relative_velocity_kmph from asteriods a join close_approach b on a.id=b.neo_reference_id ORDER BY b.relative_velocity_kmph desc Limit 10")
cursors.fetchall()
```

```python
#4. Find potentially hazardous asteriods that have approchd earth more than 3 times:

cursors.execute("select a.id,a.name,count(*) from asteriods a join close_approach b on a.id=b.neo_reference_id where a.is_potentially_hazardous_asteroid=1 group by a.id,a.name having count(*)>3")
cursors.fetchall()
```

```python
#5. find the month with the most asteriod approches:

cursors.execute("select month(close_approach_date),count(*) from close_approach group by month(close_approach_date) ORDER BY count(*) DESC Limit 1")
cursors.fetchall()
```

```python
#6. get the asteriod with the fastest ever approch speed:
cursors.execute("select a.id,a.name,b.relative_velocity_kmph from asteriods a join close_approach b on a.id=b.neo_reference_id ORDER BY b.relative_velocity_kmph desc Limit 1")
cursors.fetchall()
```

```python
#7. sort asteriods by maximum estimated diameter desc

cursors.execute("select id,name,estimated_diameter_max_km from asteriods order by estimated_diameter_max_km desc LIMIT 5")
cursors.fetchall()
```

```python
#8. an asteriod whose closest approach is getting neaerer over time (hint:use ORDERBY close_approach_date and look at miss_distance):
cursors.execute("select a.id,a.name,b.close_approach_date,b.miss_distance_km from asteriods a join close_approach b on a.id=b.neo_reference_id order by b.miss_distance_km asc Limit 5")
cursors.fetchall()
```

```python
#10. list names of asteriod that apporached earth with velocity >50000km/h

cursors.execute("""select a.name from asteriods a join close_approach b on a.id=b.neo_reference_id
 where relative_velocity_kmph>50000 Limit 5""")
cursors.fetchall()
```

```python
#11. count how many approaches happened per month:
cursors.execute("select month(close_approach_date),count(*) from close_approach group by month(close_approach_date) ORDER BY month(close_approach_date)")
cursors.fetchall()
```

```python
#12 find asteriod with the highest brightness (lowest magnitude value)
cursors.execute("select id,name,absolute_magnitude_h from asteriods order by absolute_magnitude_h asc limit 1")
cursors.fetchall()
```

```python
#13 get number of hazardous and non hazordous asteriods
cursors.execute("select is_potentially_hazardous_asteroid,count(*) from asteriods group by is_potentially_hazardous_asteroid")
cursors.fetchall()
```

```python
# 14 find asteriods that passed closer than the moon (lesser than 1ld ) along
# with their close approach date and distance
cursors.execute("""select a.name,b.close_approach_date,b.miss_distance_lunar from asteriods a join close_approach b on a.id=b.neo_reference_id where miss_distance_lunar<1 Limit 5""")
cursors.fetchall()
```

```python
#15 find asteriods that came within 0.05 AU (astronimical distance)
cursors.execute("select a.id,a.name from asteriods a join close_approach b on a.id=b.neo_reference_id where astronomical <0.05 Limit 5")
cursors.fetchall()
```

```python
%%writefile app.py
import streamlit as st
import mysql.connector
from datetime import datetime


# Initialize connection to the MySQL database
conn = mysql.connector.connect(
    user='9VyJVkmg4qyiybh.root',
    password='pbafnAX5K0Op4x5U',
    host='gateway01.ap-southeast-1.prod.aws.tidbcloud.com',
    database='Nasa_Asteriod_data'
)
cursor = conn.cursor()

# Streamlit UI setup
st.set_page_config(layout='wide')

# Title and intro
st.title("NASA NEO Tracking & Insights Dashboard")
st.markdown("""
Explore asteroid data, approach speeds, distances, and hazard insights by SQL-powered queries.
""")

# Helper function to run and display SQL queries
def show_query(query):
    cursor.execute(query)
    result = cursor.fetchall()
    st.write(result)


# Sidebar filters
st.sidebar.header("📊 Query & Filters")
query_option = st.sidebar.selectbox("Select a Query", [
    "1. Count asteroid approaches",
    "2. Average velocity per asteroid",
    "3. Top 10 fastest asteroids",
    "4. Hazardous asteroids > 3 approaches",
    "5. Month with most approaches",
    "6. Fastest ever approach",
    "7. Sort by max estimated diameter",
    "8. Closest approach getting nearer over time",
    "9. Closest approach date & distance",
    "10. Velocity > 50,000 km/h",
    "11. Approaches per month",
    "12. Brightest asteroid (lowest magnitude)",
    "13. Hazardous vs Non-hazardous count",
    "14. Asteroids < 1 LD",
    "15. Asteroids < 0.05 AU"
])

# All Queries defined here -- Chatgpt suggestions i don't know what it means
queries = {
    "1. Count asteroid approaches": ''' select neo_reference_id,count(*) from close_approach group by neo_reference_id''',
    "2. Average velocity per asteroid": ''' select neo_reference_id,avg(relative_velocity_kmph) from close_approach group by neo_reference_id''',
    "3. Top 10 fastest asteroids": ''' select a.id,a.name,b.relative_velocity_kmph from asteriods a join close_approach b on a.id=b.neo_reference_id ORDER BY b.relative_velocity_kmph desc Limit 10''',
    "4. Hazardous asteroids > 3 approaches": ''' select a.id,a.name,count(*) from asteriods a join close_approach b on a.id=b.neo_reference_id where a.is_potentially_hazardous_asteroid=1 group by a.id,a.name having count(*)>3''',
    "5. Month with most approaches": '''select month(close_approach_date),count(*) from close_approach group by month(close_approach_date) ORDER BY count(*) DESC Limit 1''',
    "6. Fastest ever approach": ''' select a.id,a.name,b.relative_velocity_kmph from asteriods a join close_approach b on a.id=b.neo_reference_id ORDER BY b.relative_velocity_kmph desc Limit 1''',
    "7. Sort by max estimated diameter": '''select id,name,estimated_diameter_max_km from asteriods order by estimated_diameter_max_km desc LIMIT 5''',
    "8. Closest approach getting nearer over time": '''select a.id,a.name,b.close_approach_date,b.miss_distance_km from asteriods a join close_approach b on a.id=b.neo_reference_id order by b.miss_distance_km asc Limit 5''',
    "9. Closest approach date & distance": '''select id,name,estimated_diameter_max_km from asteriods order by estimated_diameter_max_km desc LIMIT 5''',
    "10. Velocity > 50,000 km/h": '''select a.name from asteriods a join close_approach b on a.id=b.neo_reference_id where relative_velocity_kmph>50000 Limit 5''',
    "11. Approaches per month": '''select month(close_approach_date),count(*) from close_approach group by month(close_approach_date) ORDER BY month(close_approach_date)''',
    "12. Brightest asteroid (lowest magnitude)": '''select id,name,absolute_magnitude_h from asteriods order by absolute_magnitude_h asc limit 1''',
    "13. Hazardous vs Non-hazardous count": '''select is_potentially_hazardous_asteroid,count(*) from asteriods group by is_potentially_hazardous_asteroid''',
    "14. Asteroids < 1 LD": '''select a.name,b.close_approach_date,b.miss_distance_lunar from asteriods a join close_approach b on a.id=b.neo_reference_id where miss_distance_lunar<1 Limit 1''',
    "15. Asteroids < 0.05 AU": '''select a.id,a.name from asteriods a join close_approach b on a.id=b.neo_reference_id where astronomical <0.05 Limit 1'''
  }

# Run selected query
st.subheader(query_option)
show_query(queries[query_option])

# 🔍 Advanced Filters Section
st.header("📌 Filter Asteroid Approaches")

# User input widgets
selected_date = st.date_input("Select Close Approach Date (after)", datetime(2015, 1, 1))
min_au = st.slider("Minimum Astronomical Units (AU)", 0.0, 1.0, 0.0, 0.01)
max_au = st.slider("Maximum Astronomical Units (AU)", 0.0, 1.0, 0.05, 0.01)
min_ld = st.slider("Minimum Lunar Distance (LD)", 0.0, 100.0, 0.0, 1.0)
max_ld = st.slider("Maximum Lunar Distance (LD)", 0.0, 100.0, 10.0, 1.0)
min_velocity = st.slider("Minimum Relative Velocity (km/h)", 0.0, 100000.0, 0.0, 1000.0)
max_velocity = st.slider("Maximum Relative Velocity (km/h)", 0.0, 100000.0, 50000.0, 1000.0)
min_diameter = st.slider("Minimum Estimated Diameter (km)", 0.0, 50.0, 0.0, 0.1)
max_diameter = st.slider("Maximum Estimated Diameter (km)", 0.0, 50.0, 5.0, 0.1)
hazardous = st.selectbox("Hazardous?", ["Both", "Yes", "No"])

# Filter query
filter_query = f'''
SELECT a.name, b.close_approach_date, b.relative_velocity_kmph, b.miss_distance_km, b.miss_distance_lunar,
       a.estimated_diameter_min_km, a.estimated_diameter_max_km, a.is_potentially_hazardous_asteroid
FROM close_approach b
JOIN asteriods a ON b.neo_reference_id = a.id
WHERE date(b.close_approach_date) >= date('{selected_date}')
  AND b.miss_distance_km BETWEEN {min_au} AND {max_au}
  AND b.miss_distance_lunar BETWEEN {min_ld} AND {max_ld}
  AND b.relative_velocity_kmph BETWEEN {min_velocity} AND {max_velocity}
  AND a.estimated_diameter_max_km BETWEEN {min_diameter} AND {max_diameter}
'''
if hazardous == "Yes":
    filter_query += " AND a.is_potentially_hazardous_asteroid = 1"
elif hazardous == "No":
    filter_query += " AND a.is_potentially_hazardous_asteroid = 0"

st.subheader("Filtered Results")
show_query(filter_query)
```
